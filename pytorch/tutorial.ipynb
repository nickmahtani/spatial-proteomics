```python
# Logistic regression classifier

import torch.nn.functional as F

y = torch.tensor([1.0]) # true label
x1 = torch.Tensor([1.1]) # infput feature
w1 = torch.Tensor([2.2]) # weight parameter
b = torch.tensor([0.0]) # bias unit

z = x1 * w1 + b # net input
a = torch.sigmoid(z) # activation and output

loss = F.binary_cross_entropy(a,y)
print(loss)
```

```python
# Using the autograd engine

from torch.autograd import grad

y = torch.tensor([1.0]) # true label
x1 = torch.tensor([1.1]) # infput feature
w1 = torch.tensor([2.2], requires_grad=True) # weight parameter
b = torch.tensor([0.0], requires_grad=True) # bias unit

z = x1 * w1 + b # net input
a = torch.sigmoid(z) # activation and output

loss = F.binary_cross_entropy(a,y)

grad_L_w1 = grad(loss, w1, retain_graph=True)
grad_L_b = grad(loss, b, retain_graph=True)
print(grad_L_b)
print(grad_L_w1)
```

```python
# using high level .backward

loss.backward()

```

```python
print(w1.grad)
print(b.grad)
```

```python
# Creating a multilayer perceptron neural network

class NeuralNetwork(torch.nn.Module):
    def __init__(self, num_inputs, num_outputs):
        super().__init__()

        self.layers = torch.nn.Sequential(

            # 1st hidden layer
            torch.nn.Linear(num_inputs, 30),
            torch.nn.ReLU(),

            # 2nd hidden layer
            torch.nn.Linear(30, 20),
            torch.nn.ReLU(),

            # output layer
            torch.nn.Linear(20, num_outputs),
        )
    
    def forward(self, x):
        logits = self.layers(x)
        return logits
```

```python
torch.manual_seed(123) # to make things reproducible with the tutorial

model = NeuralNetwork(50, 3)
print(model)
```

```python
# Checking the number of trainable parameters in the model

num_params = sum(
    p.numel() for p in model.parameters() if p.requires_grad
)
print("Total number of trainable model parameters:", num_params)

```

```python
# CHecking out the weight parameter matrix of the first linear layer:

print(model.layers[0].weight) # can also use .shape
```

```python
# Accessing the bias vector

model.layers[0].bias
```

```python
# Using the neural network using a forward pass
torch.manual_seed(123)
X= torch.rand(1, 50)
out = model(X)
print(out)
```

```python
# if I want to use the model for prediction/inference

with torch.no_grad():
    out = model(X)
print(out)
```

```python
# Computing class-membership probabilities

with torch.no_grad():
    out = torch.softmax(model(X), dim=1)
print(out)
```

### 6. Using Data Loaders

```python
# Creating a toy dataset with 5 training examples with two features each

X_train = torch.tensor([
    [-1.2, 3.1],
    [-0.9, 2.9],
    [-0.5, 2.6],
    [2.3, -1.1],
    [2.7, -1.5]
])

y_train = torch.tensor([0, 0, 0, 1, 1])

X_test = torch.tensor([
    [-0.8, 2.8],
    [2.6, -1.6],
])

y_test = torch.tensor([0, 1])
```

```python
# Creating a custom dataset class

from torch.utils.data import Dataset

class ToyDataset(Dataset):
    def __init__(self, X, y):
        self.features = X
        self.labels = y

    def __getitem__(self, index):
        one_x = self.features[index]
        one_y = self.labels[index]
        return one_x, one_y
    
    def __len__(self):
        return self.labels.shape[0]

train_ds = ToyDataset(X_train, y_train)
test_ds = ToyDataset(X_test, y_test)
```

```python
len(train_ds)
```

```python
# using DataLoader class to sample from the ToyDataset

from torch.utils.data import DataLoader

torch.manual_seed(123)

train_loader = DataLoader(
    dataset = train_ds,
    batch_size=2,
    shuffle=True,
    num_workers=0,
    drop_last=True  # since 5/2 is not an int
)

test_loader = DataLoader(
    dataset=test_ds,
    batch_size=2,
    shuffle=False,
    num_workers=0
)

for idx, (x,y) in enumerate(train_loader):
    print(f"Batch {idx+1}: ", x, y)
```
